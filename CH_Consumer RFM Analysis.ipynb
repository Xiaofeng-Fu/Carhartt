## Consumer-RFM Analysis

### 1. Data Processing

#### RFM_Frequency

```python
trans["ActualDate"]=pd.to_datetime(trans["ActualDate"], errors='coerce')

consumer_freq=trans.groupby(["ConsumerKey"])["ActualDate"].nunique().to_frame('buytimes') #calculate abs buy times

trans.ActualDate.max() #the last date of the transaction record
consumer_freq["FirstDate"]=trans.groupby(["ConsumerKey"])["ActualDate"].min() #the first date of each consumers's shopping
consumer_freq["Length"]=((trans.ActualDate.max()-consumer_freq["FirstDate"])/np.timedelta64(1, 'M')).astype(float) #calculate time length in months

consumer_freq["Frequency"]=(consumer_freq.buytimes/consumer_freq["Length"]).replace(np.inf,0)  
#frequency=buy times/time length. "replace(np.inf, 0)" is to avoid the brand new customer who made their first purchase on the last date 2021-12-31.

consumer_freq=np.round(consumer_freq, decimals=2)
consumer_freq.max()
```

```python
consumer_freq
```

```python
trans
```

#### RFM_Monetary

```python
consumer_M=trans.groupby(["ConsumerKey"])['SumActualDollars'].sum().to_frame('TotalSpend')
consumer_M['buytimes']=consumer_freq.buytimes
consumer_M['Monetary']=consumer_M['TotalSpend']/consumer_M['buytimes'] #monetary=avg spending=total spend/buy times
consumer_M=np.round(consumer_M,2)
```

```python
consumer_M
```

#### RFM_Recency

```python
consumer_R=trans.groupby(["ConsumerKey"])["ActualDate"].max().to_frame("LastDate")
consumer_R['Recency']=((trans.ActualDate.max()-consumer_R["LastDate"])/np.timedelta64(1, 'M')).astype(float)
consumer_R=np.round(consumer_R,2)
```

### 2. Segmentation Methods

#### 2.1 Scoring and Binning

```python
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
RFM.isnull().sum() #check if there's NA value
```

```python
# Create labels for MonetaryValue
m_labels = range(1,6)
# Assign these labels to three equal percentile groups 
m_groups = pd.qcut(RFM['Monetary'], q=5, labels=m_labels)
# Create new column M
RFM = RFM.assign(M_Score = m_groups.values.astype(int))
RFM.head()
```

```python
#Calculating R and F groups
# Create labels for Recency and Frequency
r_labels = range(5,0,-1); f_labels = range(1,6)
# Assign these labels to 4 equal percentile groups 
r_groups = pd.qcut(RFM['Recency'], q=5, labels=r_labels)
# Assign these labels to 4 equal percentile groups 
f_groups = pd.qcut(RFM['Frequency'], q=5, labels=f_labels)
# Create new columns R and F 
RFM = RFM.assign(R_Score= r_groups.values.astype(int), F_Score = f_groups.values.astype(int))
RFM.head()
```

```python
# Concating the RFM quartile values to create RFM Segments
def join_rfm(x): return str(x['R_Score']) + str(x['F_Score']) + str(x['M_Score'])
RFM['RFM_Segment'] = RFM.apply(join_rfm, axis=1)
RFM.head()
```

```python
# Define consumer segmentation function
def consumer_segment(df):
    if ((df['R_Score']>=4) and (df['F_Score']>=3) and (df['M_Score']>=3)) :
        return 'Champions'
    elif ((df['R_Score']>=4) and (df['F_Score']>=3) and (df['M_Score']<3)) :
        return 'Loyal Customers'
    elif ((df['R_Score']>=2) and (df['R_Score']<4) and (df['F_Score']>=3) and (df['M_Score']>=3)) :
        return 'Potential Loyalists'
    elif ((df['R_Score']>=4) and (df['F_Score']<3) and (df['M_Score']>=3)):
        return 'New Customers'
    elif ((df['R_Score']>=4) and (df['M_Score']<3)and (df['M_Score']<3)):
        return 'Promising'
    elif (((df['R_Score']>=2) and (df['R_Score']<4) and (df['F_Score']<3) and (df['M_Score']>=3)) or 
          ((df['R_Score']>=2) and (df['R_Score']<4) and (df['F_Score']>=3) and (df['M_Score']<3))):
        return 'Need Attention'
    elif ((df['R_Score']>=2) and (df['R_Score']<4) and (df['F_Score']<3) and (df['M_Score']<3)):
        return 'About to Sleep'
    elif ((df['R_Score']<2) and (df['F_Score']>=3) and (df['M_Score']<3)):
        return 'At Risk'
    elif ((df['R_Score']<2) and (df['F_Score']>=3) and (df['M_Score']>=3)):
        return 'Cant Lose Them'
    elif ((df['R_Score']<2) and (df['F_Score']<3) and (df['M_Score']>=3)):
        return 'Hibernating'
    elif ((df['R_Score']<2) and (df['F_Score']<3)and (df['M_Score']<3)):
        return 'Lost'
    else:
        return 'Others'
    
          
    
# Create a new variable
RFM['Consumer_Segment'] = RFM.apply(consumer_segment, axis=1)
RFM
```

```python
# Calculate average values for each RFM_Level, and return a size of each segment 
Segment_Stat=RFM.groupby(['Consumer_Segment']).agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': ['mean', 'count']
}).round(1)
Segment_Stat
```

```python
#pip install squarify
import squarify
Segment_Stat.columns = ['RecencyMean','FrequencyMean','MonetaryMean', 'Count']
#Create our plot and resize it.
fig = plt.gcf()
ax = fig.add_subplot()
fig.set_size_inches(16, 9)
squarify.plot(sizes=Segment_Stat['Count'],label=Segment_Stat.index, alpha=.6,text_kwargs={'fontsize':13})

plt.title("RFM Segments",fontsize=18,fontweight="bold")
plt.axis('off')
plt.show()
```

#### 2.2 K-Means Clustering

```python
from mpl_toolkits.mplot3d import Axes3D

sns.set_style("white")
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(RFM.Recency, RFM.Frequency, RFM.Monetary, c='blue', s=60)
ax.view_init(30, 185)
plt.xlabel("Recency")
plt.ylabel("Frequency")
ax.set_zlabel('Monetary')
plt.show()
```

```python
from sklearn.cluster import KMeans

wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(RFM.iloc[:,:3])
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))    
plt.grid()
plt.plot(range(1,11),wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("WCSS")
plt.show()
```

```python
km = KMeans(n_clusters=6)
clusters = km.fit_predict(RFM.iloc[:,:3])

RFM["KMeans_Label"] = clusters

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(RFM.Recency[RFM["KMeans_Label"] == 0], RFM.Frequency[RFM["KMeans_Label"] == 0], RFM.Monetary[RFM["KMeans_Label"] == 0], c='blue', s=60)
ax.scatter(RFM.Recency[RFM["KMeans_Label"] == 1], RFM.Frequency[RFM["KMeans_Label"] == 1], RFM.Monetary[RFM["KMeans_Label"] == 1], c='red', s=60)
ax.scatter(RFM.Recency[RFM["KMeans_Label"] == 2], RFM.Frequency[RFM["KMeans_Label"] == 2], RFM.Monetary[RFM["KMeans_Label"] == 2], c='green', s=60)
ax.scatter(RFM.Recency[RFM["KMeans_Label"] == 3], RFM.Frequency[RFM["KMeans_Label"] == 3], RFM.Monetary[RFM["KMeans_Label"] == 3], c='blue', s=60)
ax.scatter(RFM.Recency[RFM["KMeans_Label"] == 4], RFM.Frequency[RFM["KMeans_Label"] == 4], RFM.Monetary[RFM["KMeans_Label"] == 4], c='red', s=60)
ax.scatter(RFM.Recency[RFM["KMeans_Label"] == 5], RFM.Frequency[RFM["KMeans_Label"] == 5], RFM.Monetary[RFM["KMeans_Label"] == 5], c='red', s=60)

ax.view_init(30, 185)
plt.xlabel("Recency")
plt.ylabel("Frequency")
ax.set_zlabel('Monetary')
plt.show()
```

**2) Time spent on each page:**

```python
#rearrange the ClickStream Table so that the records are in order by individual and timeline.
click_sorted=click.sort_values(['date_time','LoyaltyID'],ascending=[True,True],ignore_index=True)
click_sorted
```

```python
click_sorted['date_time']=pd.to_datetime(click_sorted["date_time"], errors='coerce')
```

```python
time_list=[['page_diff','time_diff','purchase']]
for index,row in click_sorted.iterrows():
    if (index==0) or (click_sorted.loc[index-1,'LoyaltyID']!=row['LoyaltyID']):
        min_time=row['date_time']
        min_pagenum=row['visit_page_num']
        purchase=False
    elif (pd.isnull(row['CustomerOrderNumber'])==False):
        purchase=True
        max_time=row['date_time']
        max_pagenum=row['visit_page_num']
        page_diff=max_pagenum-min_pagenum
        time_diff=max_time-min_time
        time_list.append([page_diff,time_diff,purchase])  
        #then update as a new start
        min_time=row['date_time']
        min_pagenum=row['visit_page_num']
        purchase=False
        
    elif (index==14844956) or (click_sorted.loc[index+1,'LoyaltyID']!=row['LoyaltyID']): 
        max_time=row['date_time']
        max_pagenum=row['visit_page_num']
        page_diff=max_pagenum-min_pagenum
        time_diff=max_time-min_time
       
        if (pd.isnull(row['CustomerOrderNumber'])==False):
            purchase=True
        
        if (page_diff>0) and (time_diff.total_seconds()>0):
            time_list.append([page_diff,time_diff,purchase])   
        
        
```

```python
time_df=pd.DataFrame(time_list[1:],columns=time_list[0])
time_df
```

```python
time_buy=time_df[time_df['purchase']==True]

total_login=time_df.page_diff.count()
buy_login=time_buy.page_diff.count()

percentage="{:.2%}".format(buy_login/total_login)
print('Out of a total of', total_login, 'logins,', percentage, 'of them resulted in a purchase.')
```

```python
avg_time=np.round((time_df.time_diff.sum()/time_df.page_diff.sum())/np.timedelta64(1, 's'),decimals=2)
avg_time_buy=np.round((time_buy.time_diff.sum()/time_buy.page_diff.sum())/np.timedelta64(1, 's'),decimals=2)

print('In general, people spent', avg_time, 's on viewing each page. And for people who made a purchase, they spent', avg_time_buy, 's on viewing each page.')
```
